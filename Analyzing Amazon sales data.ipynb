```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Amazon Sales data (1).csv')
```

```python
print(df.head())
```

```python
print(df.info())
print(df.describe())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Month'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Month')['Total Revenue'].sum()
```

```python
df['Year'] = df['Order Date'].dt.year
yearly_sales = df.groupby('Year')['Total Revenue'].sum()
```

```python
df['YearMonth'] = df['Order Date'].dt.to_period('M')
yearly_monthly_sales = df.groupby('YearMonth')['Total Revenue'].sum()
```

```python
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
yearly_sales.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.title('Yearly Sales Trend')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
yearly_monthly_sales.plot(kind='line', marker='o')
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue')
plt.title('Yearly-Monthly Sales Trend')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.close('all')
```
